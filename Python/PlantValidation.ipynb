```python
# Library import

import warnings

warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

import pandas as pd
from pathlib import Path
import datetime as dt
import numpy as np


from pydataxm import *                           #Se realiza la importación de las librerias necesarias para ejecutar                        
from pydataxm.pydataxm import ReadDB as apiXM 
```

```python
# Ruta General
sRutaDesp=r'C:\Alejo\cops\Python\Utils\\'
d_Fecha='2025-05-10'
#Cambiar a formato fecha
d_Fecha=dt.datetime.strptime(d_Fecha,'%Y-%m-%d')

ano=d_Fecha.year
mes=d_Fecha.month
dia=d_Fecha.day

filetype='dDEC'

l_colGen=['Planta'] 
for i in range(1,25): 
    l_colGen.append(i)
```

```python
folderpath=Path(sRutaDesp)
file=filetype + "{:02d}".format(mes) + "{:02d}".format(dia) + '.txt'
s_filepath=folderpath.joinpath(str(ano) + '-' + "{:02d}".format(mes),file)

df_Aux = pd.read_csv(s_filepath, sep=',',names=l_colGen,encoding="ISO-8859-1")
```

```python
objetoAPI = pydataxm.ReadDB()                    #Se almacena el servicio en el nombre objetoAPI
df =objetoAPI.get_collections()   
FechaIni=dt.date(2025, 5, 10)
FechaFin=dt.date(2025, 5, 10)
df_Rec= apiXM.request_data(pydataxm.ReadDB(),    #Se indica el objeto que contiene el serivicio
                        "ListadoRecursos",                #Se indica el nombre de la métrica tal como se llama en el campo metricID
                        "Sistema",                 #Campo que indica si es Sistema, Agente, Recurso, Comercializador
                        FechaIni,       #Corresponde a la fecha inicial de la consulta
                        FechaFin)      #Corresponde a la fecha final de la consulta
df_Rec=df_Rec.drop('Date',axis=1)
```

```python
df_Rec.head(5) #Se visualizan las primeras 5 filas del dataframe
```

```python
# Total plantas y mapeo de plantas en el despacho - plantas sinergox
df_Plantas = df_Aux[['Planta']]
df_Plantas=df_Plantas.merge(df_Rec, how='left', left_on='Planta', right_on='Values_Name')[['Planta','Values_Type','Values_Disp']]
df_Plantas
# df_Plantas_missing = df_Plantas[df_Plantas['Values_Type'].isna()]
# df_Plantas_missing
```

```python
# Plantas térmicas con despacho central
df_Termica=df_Plantas[(df_Plantas['Values_Type']=='TERMICA') & (df_Plantas['Values_Disp']=='DESPACHADO CENTRALMENTE')]
df_Termica=df_Termica.sort_values(by='Planta')
df_Termica.to_csv('PlantasTérmicas.csv')
df_Termica
```

```python
# Plantas hidráulicas con despacho central
df_Hidro=df_Plantas[(df_Plantas['Values_Type']=='HIDRAULICA') & (df_Plantas['Values_Disp']=='DESPACHADO CENTRALMENTE')]
df_Hidro=df_Hidro.sort_values(by='Planta')
df_Hidro.to_csv('PlantasHidro.csv')
df_Hidro
```

```python
# Plantas Eólicas y Solares con despacho central
df_Otros=df_Plantas[(~df_Plantas['Values_Type'].isin(['HIDRAULICA','TERMICA'])) & (df_Plantas['Values_Disp']=='DESPACHADO CENTRALMENTE')]
df_Otros=df_Otros.sort_values(by='Planta')
df_Otros.to_csv('PlantasOtros.csv')
df_Otros
```
