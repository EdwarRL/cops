```python
def getMes(Fecha):
    # Convert to datetime object
    oDate = dt.datetime.strptime(Fecha,'%Y-%m-%d')

    # Get month as number
    month_number = oDate.month
    iyear=oDate.year
    # print("Month number:", month_number)

    if month_number == 1:
        sMes='ene.'
    elif month_number == 2:
        sMes='feb.'
    elif month_number == 3:
        sMes='mar.'
    elif month_number == 4:
        sMes='abr.'
    elif month_number == 5:
        sMes='may.'
    elif month_number == 6:
        sMes='jun.'
    elif month_number == 7:
        sMes='jul.'
    elif month_number == 8:
        sMes='ago.'
    elif month_number == 9:
        sMes='sept.'
    elif month_number == 10:
        sMes='oct.'
    elif month_number == 11:
        sMes='nov.'
    elif month_number == 12:
        sMes='dic.'
    else:
        sMes='No se encontró el mes'


    return sMes, month_number, iyear
```
