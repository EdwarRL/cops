```python
from pathlib import Path

# ==== CONFIGURACIÓN DEL USUARIO ====
py_file_path = 'ms_modelo/app/SPED.py'            # Ruta del archivo .py generado

insert_line_number1 = 70                     # Número de línea donde insertar la nueva línea
line_to_insert1 = "def run_model(sFecha: str, iEsc: int, sTipoMod: str):\n"

insert_line_number2 = 130                     # Número de línea donde insertar la nueva línea
line_to_insert2 = "return FOVal , Status \n"

# 2. Indentar líneas
indent_start = 72  # línea inicial para indentar (inclusive)
indent_end = 136    # línea final para indentar (inclusive)
indentation = '\t'  # puede ser '\t' o '    ' (4 espacios)

# 3. Eliminar bloque
delete_start = 65  # línea inicial del bloque a eliminar
delete_end = 68    # línea final del bloque a eliminar

# ==== PROCESAMIENTO ====
# Leer archivo original
with open(py_file_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()

# Paso 1: Insertar línea 1
if 0 <= insert_line_number1 <= len(lines):
    lines.insert(insert_line_number1, line_to_insert1)
    print(f"✅ Línea insertada en la línea {insert_line_number1 + 1}.")


# Paso 2: Indentar bloque
for i in range(indent_start, indent_end + 1):
    if i < len(lines):
        lines[i] = indentation + lines[i]
print(f"✅ Líneas {indent_start + 1} a {indent_end + 1} indentadas.")

# Paso 3: Eliminar bloque de líneas
if delete_start <= delete_end and delete_start < len(lines):
    del lines[delete_start:delete_end + 1]
    print(f"✅ Bloque eliminado: líneas {delete_start + 1} a {delete_end + 1}.")

# Guardar archivo actualizado
with open(py_file_path, 'w', encoding='utf-8') as f:
    f.writelines(lines)


# Línea a insertar (puedes personalizar el contenido)
line_to_append = "\n"

# Número de líneas a insertar
num_lines = 2

# ==== PROCESO ====
# Abrir el archivo en modo 'append' para escribir al final
with open(py_file_path, 'a', encoding='utf-8') as f:
    for _ in range(num_lines):
        f.write(line_to_append)
    f.write(indentation + line_to_insert2)






print(f"🎉 Archivo actualizado correctamente: {py_file_path}")
```
