```python
def join_text_files(file_list, output_file,filepath):
    with open(output_file, 'w') as outfile:
        for file_path in file_list:
            file_path=s_mainpath.joinpath(filepath,file_path)
            if filepath:
                try:
                    with open(file_path, 'r') as infile:
                        outfile.write(infile.read())
                        outfile.write("\n")  # Add a newline character to separate files
                except FileNotFoundError:
                    print(f"File {file_path} not found.")
    print(f"All text files have been combined into {output_file}")


# Example usage
s_folderpath='Modules'
filepath=s_mainpath.joinpath(s_folderpath)
file_list = [
    '0_Setting.gms',
    '1_Sets.gms',
    '2_Parameters.gms',
    '3_Read_Data.gms',
    '4_Variables.gms',
    '5_PreProcess.gms',
    '6_Equation_Definition.gms',
    '7_Equations.gms',
    '8_Model_Execution.gms',
    '9_PostProcess.gms',
    '10_WriteProcess.gms'
  # Add paths to all the text files you want to join

]

output_file=s_mainpath.joinpath('joinfile.txt')

join_text_files(file_list, output_file,filepath)
```

```python
import re

def find_and_print_matched_lines(input_file, output_file, l_espace):
    """
    Encuentra líneas que comienzan con un patrón específico y terminan con otro patrón
    en un archivo de entrada, y las imprime en otro archivo de salida.

    :param input_file: Ruta del archivo de entrada.
    :param output_file: Ruta del archivo de salida donde se imprimirán las líneas coincidentes.
    :param start_pattern: Patrón de expresión regular para el inicio de la línea.
    :param end_pattern: Patrón de expresión regular para el final de la línea.
    """
    # Construir el patrón de expresión regular
    regex = re.compile("^^----  .* Equation")
    regex1 = re.compile("^^----  .* Assignment")
    
    # Abrir archivo de entrada para lectura y archivo de salida para escritura
    with open(input_file, 'r', encoding='utf-8') as fin, open(output_file, 'w', encoding='utf-8') as fout:
        for line in fin:
            # Buscar el patrón en la línea actual
            if regex.search(line) or regex1.search(line):

                # Dividir la línea en palabras
                words = line.strip().split()

                formatted_line=''
                for icont in range(0,len(words)):
                    # Aplicar el espaciado entre las últimas palabras
                    ispace=l_espace[icont]-len(words[icont])
                    formatted_line = formatted_line + words[icont] + ' ' * ispace

                # Escribir la línea en el archivo de salida
                fout.write(formatted_line + '\n')

# Ejemplo de uso
input_file = "Main_COPS.lst"    # Ruta del archivo de entrada
output_file = "ExecutionTimes.txt"  # Ruta del archivo de salida
l_espace=[8,10,15,35,10,10,10,10,10,10]

# Encontrar líneas que coinciden con el patrón de inicio y fin en el archivo de entrada, y escribirlas en el archivo de salida
find_and_print_matched_lines(input_file, output_file, l_espace)
```
